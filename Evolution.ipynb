```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
from math import sqrt
```

```python
N, sigma, p, m, h, t = smp.symbols('N, sigma, p, m, h, t', real = True, positive = True)
x = smp.symbols('x', real = True)
```

```python
def psi(x, t, m = 1, h = 1, N = 1, sigma = 1):
    return sqrt(2 * m) * N * sigma * (1 + 1j) * np.exp(- (m * pow(x, 2)) / (2 * (1j * h * t + m * pow (sigma, 2)))) / (1j * m * pow(sigma, 2) - h * t) ** 0.5 / 2
```

```python
plt.style.use('Solarize_Light2')
mpl.rcParams['font.size'] = 30       
plt.figure(figsize = (20, 20))

plt.title(r'$|\psi(x)|$, N = 1, sigma = 1, m = 1, h = 1', fontsize = 30)
plt.xlabel(r"x")
plt.ylabel(r'$|\psi(x)|$')

x = np.linspace(-50, 50 , 500)

for i in range (500):
    plt.plot(x, np.abs(psi(x, i)), label=f"t = {i}")

plt.tight_layout()
plt.legend(loc = 'best', fontsize = 10)
plt.show()
```

Состояние локализованное, но не связанное (ВФ уходит на бесконечность)

При $\sigma \rightarrow 0$ в пределе получим зарождающуюся дельта-функцию.
